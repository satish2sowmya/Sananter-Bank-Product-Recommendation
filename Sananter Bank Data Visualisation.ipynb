**Influence of Age on Product Usage**

Let us look at the influence of age on product usage. For this analysis , let us divide customers into different 
age groups less than 15( child to high school).16-30(Undergrad,masters,Phd).30-45(establishing homes, child education,investments for future)45-60(children out of homes,children in college),60and above(retired and enjoying)

```python
df['Age_buckets']=df['Age'].apply( lambda x : np.where(x<15,1,(np.where(x<30,2,np.where(x<45,3,np.where(x<60,5,6))))))
```

```python
plt.scatter(df['Age'],df['Age_buckets'])
plt.xlabel("CUSTOMER AGE",fontsize=15,weight='bold')
plt.ylabel("AGE_BUCKET",fontsize=15,weight='bold')
plt.title("CUSTOMER AGE BUCKETS",fontsize=15,weight='bold')
```

```python
df_age=pd.DataFrame(df.groupby('Customer_code')['Age_buckets'].apply(lambda x : x.max()))
# assigning customer to his latest age bucket
```

```python
df_age_products=df_age.join(df_products) 
# creating a new data frame for age bucket and products
```

```python
fig=plt.figure(figsize=(15,5))
ax1=fig.add_subplot(1,1,1)
df_age_products.groupby('Age_buckets')[list_products].apply(lambda x : x.sum()/len(x)).unstack().unstack().plot(kind='bar',ax=ax1)
plt.xlabel("AGE_BUCKETS",fontsize='15',weight='bold')
plt.ylabel("PROPORTION",fontsize='15',weight='bold')
plt.title("PRODUCT VARIATION WITH AGE BUCKET", weight='bold')
```

Group 1(children to college going):High proportion of junior account usage.There is some usage of pay roll and pensions but need to verify 
    if it is due to error in the data
    
Group 2(college going to newly employed): High usage of curren account. Some payroll,pension and direct debit usage.
    They donot have any presence in other account types.
Group 3(middle aged),Group 4(old aged) ,group5(retired ).They have representation in almost all the products except 
junior account.
